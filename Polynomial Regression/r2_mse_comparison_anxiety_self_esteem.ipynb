```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
exercise = [0.0, 0.53, 1.05, 1.58, 2.11, 2.63, 3.16, 3.68, 4.21, 4.74,
5.26, 5.79, 6.32, 6.84, 7.37, 7.89, 8.42, 8.95, 9.47, 10.0]                                     
calories = [7.99, 9.17, 11.91, 12.89, 9.70, 11.32, 15.49, 17.53, 20.27, 20.92,
24.76, 24.80, 26.36, 26.21, 29.41, 33.71, 31.72, 36.96, 32.63, 35.60]
```

```python
plt.scatter(exercise, calories, color='red', label='Data')
```

```python
exercise_np = np.array(exercise).reshape(-1, 1)
calories_np = np.array(calories)

poly = PolynomialFeatures(degree=1)
x_poly = poly.fit_transform(exercise_np)

model = LinearRegression()
model.fit(x_poly, calories_np)
```

```python
y_pred = model.predict(x_poly)  

r2 = model.score(x_poly, calories)
print(f"R² by model.score = {r2:.4f}")

r2 = r2_score(calories, y_pred)
print(f"R² by r2_score = {r2:.4f}")

mse = mean_squared_error(calories, y_pred)
print(f"MSE: {mse:.2f}")

```

```python
x_line = np.linspace(0, 10, 20).reshape(-1, 1)
x_line_poly = poly.transform(x_line)
y_line = model.predict(x_line_poly)

plt.scatter(exercise, calories, color='red', label='Data')
plt.plot(x_line, y_line, color='blue', label='Calories Burn Line')
plt.xlabel('Weekly Exercise')
plt.ylabel("Calories Burn" )
plt.title('Exercise vs. Calories Burn')
plt.legend()
plt.grid(True)
plt.show()
```

# The Linear Model fits perfectly!

```python
anxiety_level = [-4.0, -3.58, -3.16, -2.74, -2.32, -1.89, -1.47, -1.05, -0.63, -0.21,
0.21, 0.63, 1.05, 1.47, 1.89, 2.32, 2.74, 3.16, 3.58, 4.0]
self_esteem = [41.02, 43.27, 39.14, 25.96, 23.08, 16.41, 14.31, 12.26, 1.91, 5.65,
3.80, -5.23, -10.89, -2.85, -3.41, -3.34, -0.75, 13.14, 11.03, 17.77]
```

```python
plt.scatter(anxiety_level, self_esteem, color='red', label='Data')
```

```python
anxiety_level_np = np.array(anxiety_level).reshape(-1, 1)
self_esteem_np = np.array(self_esteem)

poly = PolynomialFeatures(degree=1)
x_poly = poly.fit_transform(anxiety_level_np)

model = LinearRegression()
model.fit(x_poly, self_esteem_np)

y_pred = model.predict(x_poly)  

r2 = r2_score(self_esteem, y_pred)
print(f"R² for Linear Line = {r2:.4f}")

mse = mean_squared_error(self_esteem, y_pred)
print(f"MSE for Linear Line: {mse:.2f}")
```

```python
x_line = np.linspace(-5, 5, 20).reshape(-1, 1)
x_line_poly = poly.transform(x_line)
y_line = model.predict(x_line_poly)

plt.scatter(anxiety_level, self_esteem, color='red', label='Data')
plt.plot(x_line, y_line, color='blue', label='Self Esteem Line')
plt.xlabel('Anxiety Level')
plt.ylabel("Self Esteem" )
plt.title('Anxiety Level vs. Self Esteem')
plt.legend()
plt.grid(True)
plt.show()
```

```python
anxiety_level_np = np.array(anxiety_level).reshape(-1, 1)
self_esteem_np = np.array(self_esteem)

poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(anxiety_level_np)

model = LinearRegression()
model.fit(x_poly, self_esteem_np)

y_pred = model.predict(x_poly)  

r2 = r2_score(self_esteem, y_pred)
print(f"R² for PolyLine = {r2:.4f}")

mse = mean_squared_error(self_esteem, y_pred)
print(f"MSE for Poly Line: {mse:.2f}")
```

```python
x_line = np.linspace(-5, 5, 20).reshape(-1, 1)
x_line_poly = poly.transform(x_line)
y_line = model.predict(x_line_poly)

plt.scatter(anxiety_level, self_esteem, color='red', label='Data')
plt.plot(x_line, y_line, color='blue', label='Self Esteem Line')
plt.xlabel('Anxiety Level')
plt.ylabel("Self Esteem" )
plt.title('Anxiety Level vs. Self Esteem')
plt.legend()
plt.grid(True)
plt.show()
```

# According to the graph plots, R² values and MSE values, The PolyLine Model is obviousely more suitable to describe the relations between a person's Anxiety Level and Self Esteem
